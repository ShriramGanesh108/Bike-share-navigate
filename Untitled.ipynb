```python
all_dfs = []
for filename in glob.glob('*divvy-tripdata.csv'):
    print (f'Loading {filename}')
    new_df =pd.read_csv(filename, usecols=['ride_id',
                        'rideable_type',
                        'started_at',
                        'ended_at',
                        'start_station_name',
                        'start_station_id',
                        'end_station_name',
                        'end_station_id',
                        'start_lat',
                        'start_lng',
                        'end_lat',
                        'end_lng',
                        'member_casual'])
    all_dfs.append(new_df)
```

```python
df =pd.concat(all_dfs)
```

```python
df.shape
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df["started_at"] = df["started_at"].astype('datetime64[ns]')
```

```python
df["ended_at"] = df["ended_at"].astype('datetime64[ns]')
```

```python
df
```

```python
df["ride_time"] = df["ended_at"]-df["started_at"]
```

```python
df
```

```python
from math import radians
```

```python
df['start_lat'] =np.radians(df['start_lat'])
```

```python
df['end_lat'] =np.radians(df['end_lat'])
```

```python
df['start_lng'] =np.radians(df['start_lng'])
```

```python
df['end_lng'] =np.radians(df['end_lng'])
```

```python
df
```

```python
df['dlon'] = df['end_lng'] - df['start_lng']
df['dlat'] = df['end_lat'] - df['start_lat']
```

```python
df['a']=np.sin(df['dlat'] / 2)**2 + np.cos(df['start_lat']) * np.cos(df['end_lat']) * np.sin(df['dlon'] / 2)**2
```

```python
df['a']
```

```python
df['total_distance'] =2 * np.arcsin(np.sqrt(df['a']))
```

```python
df['total_distance']
```

```python
df['total_distance']=df['total_distance']*6376
```

```python
df['total_distance']
```

```python
df
```

```python
df
```

```python
df.drop(['dlon','dlat','a'],True,inplace =True)
```

```python
df
```

```python
pd.set_option('precision', 3)
```

```python
df
```

```python
df.drop(['start_lat','start_lng','end_lat','end_lng'],True,inplace =True)
```

```python
df['ride_time']=df['ride_time']/60
```
